```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd 
import numpy as np 

# Load dataset
data = load_breast_cancer()
X = data.data  # Features (30 columns)
y = data.target  # Target (binary: malignant/benign)
feature_names = data.feature_names

print("Original features:", X.shape[1])
```

```python
# Select top 10 features
k = 10
selector = SelectKBest(score_func=f_classif, k=k)
X_new = selector.fit_transform(X, y)

# Get selected feature names and scores
selected_features = [feature_names[i] for i in selector.get_support(indices=True)]
scores = selector.scores_[selector.get_support()]

print("Selected features:", selected_features)
print("Scores:", scores)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create DataFrame of scores
df_scores = pd.DataFrame({'Feature': feature_names, 'Score': selector.scores_})
# print("first",df_scores)
df_scores = df_scores.sort_values('Score', ascending=False)


# Plot
plt.figure(figsize=(10, 6))
plt.bar(df_scores['Feature'][:k], df_scores['Score'][:k], color='skyblue')
plt.xticks(rotation=90)
plt.title(f"Top {k} Features by ANOVA F-Score")
plt.ylabel("F-Score")
plt.show()
```
